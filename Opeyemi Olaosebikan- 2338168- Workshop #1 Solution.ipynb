* Q10: Write a function that receives the dataset and replace Famle with F and Male with M (please try to write it yourself before checking the answer in the next cell)

```python
# Answer
def encode_sex(data):
    rows=data.shape[0]
    a=data.loc[:,'sex']
    for i in range(rows): 
        if a[i]=="Male":
            data.loc[i,"sex"]="M"
        elif a[i]=="Female":
            data.loc[i,"sex"]="F"
    return data
```

## TASK 1.6: Write a function to split the task_dataset in half column-wise and swap the first half and the second half

```python
def swap_columns(task_dataset):
    # Split the dataset in half colunms
    num_colunms = len(task_dataset.columns)
    half_point = num_colunms // 2
    first_half = task_dataset.iloc[:, :half_point]
    second_half = task_dataset.iloc[:, half_point:]
    
    #swap the first and second halves
    swapped_task_dataset = pd.concat([second_half, first_half], axis=1)
    
    return swapped_task_dataset
    
swap_columns(task_dataset)
```

We divided the columns in the dataset into half and swapped positions of DataFrame halves by concatenating along columns axis.Relationship - income columns are now before age - occupation

## Task 1.8: Write a function that returns the names of countries with maximum and minimum average ages in the task_dataset (Hint: you can use the numpy module)

```python
############# WRITE THE CODE IN THIS CELL####################

def find_extreme_avg_age_countries(task_dataset):
    
    # Calculate average age for each country
    avg_age_by_country = task_dataset.groupby('native-country')['age'].mean() 
    
    # Find the country with the maximum average age
    max_avg_age_country = avg_age_by_country.idxmax() 
    
    # Find the country with the minimum average age
    min_avg_age_country = avg_age_by_country.idxmin() 

    return max_avg_age_country, min_avg_age_country

max_age_country, min_age_country = find_extreme_avg_age_countries(task_dataset)
print(f"Country with maximum average age: {max_age_country}")
print(f"Country with minimum average age: {min_age_country}")
```

The average age for each country is computed and the countries with the highest and lowest average ages identified
