# For Portugal_Online_Retails

```python
#import the dataset
portugal_basket = pd.read_csv('Portugal_online_retail.csv')
```

```python
# View the dataset
display(portugal_basket.head())
```

```python
# dropping the InvoiceNo column
portugal_basket = portugal_basket.drop('InvoiceNo',axis=1)
```

```python
#view new portugal data without invoice column
portugal_basket.head()
```

```python
# Finding the frequent itemsets in descending order
frequent_items_portugal = apriori(portugal_basket, min_support=0.09, use_colnames=True)
frequent_items_portugal.sort_values(by=['support'],ascending=False, ignore_index=True)
```

```python
# Define confidence level 
confidence_level = 0.7

# Collecting the inferred rules in a dataframe
rules_portugal = association_rules(frequent_items_portugal, metric="confidence", min_threshold=0.1)

# sort rules in descending order
rules_portugal.sort_values(by=['support'],ascending=False, ignore_index=True)
```

```python
# Select rules with the chosen confidence level
selected_rules_portugal = rules_portugal[rules_portugal['confidence'] >= confidence_level]

# Get the top three rules based on support
top_three_rules_portugal = selected_rules_portugal.sort_values(by='support', ascending=False).head(3)

# Display the top three rules
display(top_three_rules_portugal[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

The output shows the top three association rules for the Portugal dataset at the confidence_level = 0.7. The rules are displayed in a dataframe with columns for the antecedents, consequents, support, confidence, and lift.

The first rule shows that there is a strong association between the antecedent "BAKING SET 9 PIECE RETROSPOT" and the consequent "RETROSPOT TEA SET CERAMIC 11 PC". The support for this rule is 0.155, which means that 15.5% of the transactions in the dataset contain both items. The confidence is 0.75, which means that 75% of the transactions that contain the antecedent also contain the consequent. The lift is 3.11, which indicates that the association between the antecedent and consequent is 3.11 times stronger than would be expected if they were independent.

The second and third rules show that there is a strong association between the "JUMBO BAG PINK VINTAGE PAISLEY" and "JUMBO SHOPPER VINTAGE RED PAISLEY" items. The support for both rules is the same (0.155), and the confidence for the second rule is 1.0, which means that all transactions that contain the antecedent also contain the consequent. The lift for both rules is 5.27, indicating a very strong association.
