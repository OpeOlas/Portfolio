```python
##### TASK 2.2c #####

import pandas as pd

# Import the dataset again.
adult2 = pd.read_csv('adult.csv')

# Find the mode of the categorical (i.e. nominal or ordinal) columns which contain Null values and use it to fill their Null values
categorical_columns = adult2.select_dtypes(include='object').columns
null_categorical_columns = categorical_columns[adult2[categorical_columns].isna().any()]

print(null_categorical_columns)
 
```

```python
# Find the mode of each null categorical column and fill the Null values with the mode
for col in null_categorical_columns:
    mode = adult2[col].mode().iloc[0]
    adult2.loc[adult2[col].isna(), col] = mode
    
adult2.isna().sum()
    
```
